```python
# Extract the numeric part of the increment code and calculate total time
# the first the initial time and the second is the increment per move
# the number of moves is column named 'turns'
# but be careful that the first part is in minutes and the second part is in seconds, so we need to convert the second part to minutes before adding it to the initial time
def calculate_total_time(row):
	initial_time = int(row['increment_code'].split('+')[0])
	increment = int(row['increment_code'].split('+')[1]) / 60  # Convert seconds to minutes
	total_time = initial_time + (increment * row['turns'])  # Assuming an average of 40 moves per game
	return total_time
df['total_time'] = df.apply(calculate_total_time, axis=1)
```

```python
df['initial_time'] = df['increment_code'].apply(lambda x: int(x.split('+')[0]))
df['increment'] = df['increment_code'].apply(lambda x: int(x.split('+')[1]))  # Convert seconds to minutes
```

```python
def classify_game(row):
	initial_minutes = row['initial_time']
	increment_seconds = row['increment']
	if initial_minutes * 60 + increment_seconds * 60 <= 180:
		return 'Bullet'
	elif initial_minutes * 60 + increment_seconds * 60 < 600:
		return 'Blitz'
	elif initial_minutes * 60 + increment_seconds * 60 < 1800:
		return 'Rapid'
	else:
		return 'Classic'

df['type'] = df.apply(classify_game, axis=1)
```
